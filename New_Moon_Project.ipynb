```python
import yfinance as yf
import datetime
import pandas as pd
```

```python
symbol = "^BSESN"

end_date = datetime.date.today()
start_date = end_date - datetime.timedelta(days=365)  


sensex = yf.Ticker(symbol)

historical_data = sensex.history(period="1y")


```
